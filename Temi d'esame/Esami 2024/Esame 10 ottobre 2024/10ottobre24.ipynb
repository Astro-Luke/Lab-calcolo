```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Funzioni per il programma

# Funzione di sturges per il binnaggio
def sturges (N_eventi) :
    return int (np.ceil (1 + np.log2 (N_eventi)))  

# Funzione Rand Range
def rand_range (xmin, xmax) :
    return xmin + (xmax - xmin) * random.random ()

# Media con lista
def media (lista) :
    mean = sum(lista)/len(lista)
    return mean
    
# Varianza con lista
def varianza_bessel (lista) :
    somma_quadrata = 0
    for elem in lista :
        somma_quadrata = somma_quadrata + (elem - media(lista))**2
    return somma_quadrata/(len(lista) -1)

# Deviaz. standard con lista
def dev_std (lista) :
    sigma = np.sqrt(varianza_bessel(lista))
    return sigma

def dev_std_media (lista) :
    return dev_std(lista)/np.sqrt(len(lista))
```

Si scriva una funzione chiamata generate_gaus_bm che generi coppie di numeri pseudo-casuali
distribuiti secondo una densità di probabilità Gaussiana utilizzando l’algoritmo di Box-Müller, implementata in una libreria dedicata.

```python
def generate_gaus_bm () :
    x1 = rand_range (0., 1.)
    x2 = rand_range (0., 1.)
    g1 = np.sqrt (-2*np.log(x1)) * np.cos(2*np.pi*x2)
    g2 = np.sqrt (-2*np.log(x1)) * np.sin(2*np.pi*x2)
    return g1, g2
```

Si generino N = 1000 numeri pseudo-casuali utilizzando la funzione appena sviluppata e li si disegni
in un istogramma, scegliendone con un algoritmo opportuno gli estremi ed il binnaggio.

```python
N = 1000
lista_casuali = []
for i in range (N//2) :
    g1, g2 = generate_gaus_bm ()
    lista_casuali.append (g1)
    lista_casuali.append (g2)

Nbin = sturges (N)

bin_content, bin_edges = np.histogram (lista_casuali, bins=Nbin, range = (min(lista_casuali), max(lista_casuali))) 

fig, ax = plt.subplots ()
ax.hist (lista_casuali, bins=bin_edges, color = 'orange')
ax.set_title ('Istogramma Box-Mueller', size = 14)
ax.set_xlabel ('x')
ax.set_ylabel ('y')
ax.grid ()                                          # Se voglio la griglia

plt.show () 
```

Si determinino media e varianza della distribuzione ottenuta e relativi errori.

```python
mean = media (lista_casuali)
var = varianza_bessel (lista_casuali)

print ('\nMedia della distribuzione: ', mean)
print ('\nVarianza della distribuzione: ', var)
```

Si mostri graficamente che, al variare del numero N di eventi generati, la sigma della distribuzione
non cambia, mentre l’errore sulla media si riduce.

```python
N_max = 200000
lista_casual_N = []
sigma = []
sigma_mean = []
lista_N = []
count = 1

while (count < N_max) :
    g1, g2 = generate_gaus_bm ()
    lista_casual_N.append (g1)
    lista_casual_N.append (g2)
    
    sigma.append (np.std (lista_casual_N))
    sigma_mean.append (dev_std_media (lista_casual_N))

    lista_N.append (count) 
    count = count * 2

fig, ax = plt.subplots ()
ax.plot (lista_N, sigma_mean, color = 'blue', label = 'dev std media')
ax.plot (lista_N, sigma, color = 'red', label = 'dev std')
ax.set_title ('Andamento dev std vs dev std media')
ax.set_xlabel ('N eventi')
ax.set_ylabel ('y')
ax.set_xscale ('log')
#ax.set_yscale ('log')
ax.grid ()
ax.legend ()

plt.show ()
```

Si trasformi l’algoritmo in modo che generi numeri pseudo-casuali con densità di probabilità Gaussiana con media µ = 5 e varianza σ
2 = 4. Si generi un nuovo campione di N = 1000 eventi con il
nuovo algoritmo e se ne disegni la distribuzione, sempre scegliendo in modo opportuno gli estremi
ed il binnaggio dell’istogramma corrispondente.

```python
def generate_bm (mu, std) :
    g1, g2 = generate_gaus_bm ()
    new_g1 = std*g1 + mu
    new_g2 = std*g2 + mu
    return new_g1, new_g2
```

```python
new_casual = []
for i in range (N//2) :
    g1, g2 = generate_bm (5, 2)
    new_casual.append (g1)
    new_casual.append (g2)

Nbin = sturges (N)

bin_content, bin_edges = np.histogram (new_casual, bins=Nbin, range = (min(new_casual), max(new_casual))) 

fig, ax = plt.subplots ()
ax.hist (new_casual, bins = bin_edges, color = 'orange')
ax.set_title ('Istogramma con parametri gaussiani')
ax.set_xlabel ('x')
ax.set_ylabel ('y')
ax.grid ()

plt.show ()
```
